# IP Address
We use IPV4 like 192.168.0.1<br> here the pattern is like A.A.A.A where each A is of 8 bit so total size of IPV4 is 32 bit.<br>IP address = Network address(N) + Host address(H).The combination can be like <ol><li>N.N.N.H</li><li>N.N.H.H</li><li>N.H.H.H</li></ol>Now in the 8 bit of each address what's the **min & max value** min is 0 and max is 255. so it's like 0 ~ 255.0 ~ 255.0 ~ 255.0 ~ 255. But in the host address part we need to reserve two addresses one reserved for network and another one reserved for broadcast. So in 1st combination we can connect 256-2=254 devices. and the range of address should be 1 ~ 254.<br> **Ex** : 192.168.0.0 -- Network address reserved<br>192.168.0.255 -- Broadcast address reserved<br>  In the 2nd combination there are total 2 host address so total connection will be 2^16(as each is 8bit) - 2 = 65534. And in the 3rd combination we can connect 2^24 - 2 = 16million devices.


The are types of IP address called class. <ol><li>Class C (H <= 254)</li><li>Class B (H <= 65534)</li><li>Class A (H <= 16millions)</li></ol>These classes are based on host connection requirements. **Now how to understand which IP belongs to which class?**<br> To answer thi we need to look at the fisrt octect of the IP that is the first A from A.A.A.A .<br>
<ul><li>1 ~ 127 = Class A</li><li>128 ~ 191 = Class B</li><li>192 ~ 223 = Class C</li></ul>As an example 192.168.0.1 here the first octect is 192 so it belongs to class C.<br>
